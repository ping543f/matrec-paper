```python
from scidownl.scihub import *
import pandas as pd
import re

# df = pd.read_excel("carbon-papers-webofscience-acta-carbon-power_edited_jr.xlsx",usecols =['DOI'])

# doi_list = df.values.tolist()
f = open("doi-list-test-papers.txt","r")
content = f.readlines()

print(len(content))





```

```python
DOIs = []
###For Pandas using direct excel file input for DOI list
# for doi in doi_list:
#     doi_ = str(doi)
#     doi_ = re.sub("(\[)|(\])|(')","",doi_)
#     DOIs.append(doi_)

### If excel does not work, copy paste the doi columns in a text file and use that list
for doi in content:
    doi_ = str(doi)
    doi_ = re.sub("\\n","",doi_)
    DOIs.append(doi_)

print (len(DOIs))
```

```python
out = 'test-papers-edlc-related/'
counter = 1
fwrite = open("doi_download_log.txt","a+")
for doi in DOIs:
    print('****************  ',counter,'   *******************')
    print(doi)
    SciHub(doi, out).download(choose_scihub_url_index=3)
    counter +=1
    fwrite.write(doi+"\n")
fwrite.close()
print("Done!!")
```
